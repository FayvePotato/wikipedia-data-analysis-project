```python
import pandas as pd

import sqlite3
```

```python
wikipedia_data = pd.read_excel('final_clean_data.xlsx',  header=0)
```

```python
wikipedia_data.head()
```

```python
wikipedia_db_conn = sqlite3.connect('wikipedia_data.db')
```

```python
c = wikipedia_db_conn.cursor()
```

```python
c.execute("""

    CREATE TABLE page_visits(
        date DATE,
        page TEXT NOT NULL,
        visits INTEGER
    );

""" )
```

```python
wikipedia_data.to_sql('page_visits', wikipedia_db_conn, if_exists='append', index=False)
```

```python
pd.read_sql("select * from page_visits limit 10", wikipedia_db_conn)
```

question 5 (which day is the most popular) & 6 (which day is the least popular)

```python
weekdays = pd.read_sql("select date, case cast (strftime('%w', date) as integer) when 0 then 'Sunday' when 1 then 'Monday' when 2 then 'Tuesday' when 3 then 'Wednesday' when 4 then 'Thursday' when 5 then 'Friday' else 'Saturday' end as weekday from page_visits  ",  wikipedia_db_conn)

weekdays
```

```python
weekdays.to_excel('tables/weekdays.xlsx')
```
